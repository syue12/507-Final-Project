```python
import json

data_1 = json.loads(open("new_harry_potter_characters.json","r").read())
data_2 = json.loads(open("harry_potter_stats.json","r").read())
data_3 = json.loads(open("harry_potter_stats_rest.json","r").read())
```

```python
data_4 = data_2 + data_3
len(data_4)
```

```python
len(data_1)
```

```python
import pandas as pd

df_1 = pd.DataFrame.from_dict(data_1)
df_2 = pd.DataFrame.from_dict(data_4)

hp_data = df_1.join(df_2.set_index('name'), on='name')
hp_data
```

```python
affiliations = hp_data["affiliations"].tolist()
aff = [name.lower().strip() for item in affiliations for name in item.split(",")]
aff
```

```python
from collections import Counter
c = Counter(aff)
c.most_common(10)
```

```python
personality = hp_data["personality"].tolist()
personality = [name.lower().strip() for item in personality for name in item.split(",")]
personality
```

```python
from collections import Counter
c = Counter(personality)
c.most_common(15)
```

```python
hp = pd.read_csv('hp.csv', index_col=[0])  
```

```python
hp['likes'] = hp['likes'].fillna(0)
hp['comments'] = hp['comments'].fillna(0)
hp['views'] = hp['views'].fillna(0)
hp = hp.fillna("none")
hp.head()
```

```python
hp.loc[(hp["likes"] >= 100), "popularity"] = 1
hp.loc[(hp["likes"] >= 1000), "popularity"] = 2
hp.loc[(hp["likes"] >= 10000), "popularity"] = 3
hp.loc[(hp["likes"] >= 100000), "popularity"] = 4
hp.loc[(hp["likes"] >= 1000000), "popularity"] = 5
```

```python
hp['friends'] = hp['friends'].str.split(', ')
hp['enemies'] = hp['enemies'].str.split(', ')
hp['personality'] = hp['personality'].str.split(', ')
hp['affiliations'] = hp['affiliations'].str.split(', ')
```

```python
hp['personality'] = hp['personality'].apply(lambda x: [str(a).lower() for a in x])
hp['affiliations'] = hp['affiliations'].apply(lambda x: [str(a).lower() for a in x])
```

```python
hp['popularity'] = hp['popularity'].fillna(0)
```
